```python
#prompt:traduza para portugues as categorias da coluna "senioridade" do dataframe df
df['senioridade'] = df['senioridade'].replace({
    'SE': 'Senior',
    'MI': 'Medio',
    'EN': 'Junior',
    'EX': 'Executivo'
})

df['senioridade'].value_counts()
```

```python
#prompt:traduza para portugues as categorias da coluna "contrato" do dataframe df
df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Meio Periodo',
    'FL': 'Freelancer'
})

df['contrato'].value_counts()
```

```python
#prompt:traduza para portugues as categorias da coluna "remoto" do dataframe df
df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    100: 'Remoto',
    50: 'Híbrido'
})

df['remoto'].value_counts()
```

```python
#FILTRANDO DADOS NULOS UTILIZANDO A MÉDIA

#biblioteca que permite dados nulos
import numpy as np

#criando um dataframe (df_salarios) para teste com as colunas nome e salário
#np.nan->são os dados nulos
df_salarios = pd.DataFrame({
    "nome":["Lucas","Juliana","Raphael","Junior","Satie"],
    "salario":[np.nan,4500,5000,np.nan,100000]
  })

'''
criando a coluna (salario_media) e recebendo a fórmula .fillna para preencher os campos nulos
df_salarios -> base de dados
salario_media -> coluna nova
dentro da base na coluna média iremos prencher os valores nulos(fill.na) com a média ('salario'.mean())
.round2 serve para fazer o arrendodamento do valor
criando a coluna salario_mediana para prencher os campos nulos com o median
.median serve também para calcular a média porém se tiver um valor fora do padrão ele desconsidera da média
'''

df_salarios["salario_media"] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

df_salarios["salario_mediana"] = df_salarios['salario'].fillna(df_salarios['salario'].median())
df_salarios
```

```python
#ALTERANDO VALORES NULOS PELO VALOR SUPERIOR E ANTERIOR

#criando o dataframe temperatura com as colunas dia e temperatura
df_temperaturas = pd.DataFrame({
    "Dia": ['Segunda','Terça','Quarta','Quinta','Sexta','Sábado','Domingo'],
    "Temperatura": [30, np.nan, np.nan, 28, 27, np.nan, 26]
})

#criando a coluna preenchido_ffil com a funcão de caso tenha um valor nulo preenche com o valor anterior .ffill()
#criando a coluna preenchido_bfil com a função de caso tenha um valor nulo preenche com o valor o próximo valor .bfill()
df_temperaturas["preenchido_ffill"] = df_temperaturas["Temperatura"].ffill()
df_temperaturas["preenchido_bfill"] = df_temperaturas["Temperatura"].bfill()

print(df_temperaturas)
```

```python
#ALTERANDO VALORES NULOS PARA STRING 'Não informado'

#criando o dataframe df_cidades com as colunas nome e cidade
df_cidades = pd.DataFrame({
    'nome':['Lucas','Juliana','Raphael','Junior','Satie'],
    'cidade':['SP',np.nan,'BH','RJ',np.nan]
})
#adicionando ao dataframe uma coluna cidade_preenchida
#os valores nulos ficaram como "Não Informado"
df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('Não informado')
df_cidades
```
